```python
# import
import pandas as pd
import csv
```

```python
# Store filepath in a variable
file_one = "Resources/purchase_data.csv"
```

```python
# Read file
file_one_df = pd.read_csv(file_one)
```

```python
file_one_df.head()
```

```python
# Stat Overview
file_one_df.describe()
```

```python
# Total Players
total_players = len(file_one_df["SN"])
#nunique counts distinct values.

player_count = file_one_df['SN'].nunique()
player_count
total_players
```

```python
#Summary stats
TotalOrders = file_one_df["Item ID"].count()
TotalRevenue = file_one_df["Price"].sum()
AverageOrder = file_one_df["Price"].mean()
UniqueItems = file_one_df["Item Name"].nunique()

# Creating a summary DataFrame using above values
purchases_df = pd.DataFrame({"Total Order Count":[TotalOrders],
                             "Total Unique Items": [UniqueItems],
                             "Average Sale": [AverageOrder],
                           "Total Revenue": [TotalRevenue]})

purchases_df["Total Revenue"] = purchases_df["Total Revenue"].map("${:.2f}".format)
purchases_df["Average Sale"] = purchases_df["Average Sale"].map("${:.2f}".format)

purchases_df
```

```python
# Percentage and Count of Male Players and Female Players and non-disclosed

male = (file_one_df['Gender']=="Male").sum()
female = (file_one_df['Gender']=="Female").sum()
other = (file_one_df['Gender']=="Other / Non-Disclosed").sum()
pctmale = (male / total_players) * 100
pctfemale = (female / total_players)* 100
pctothers = (other / total_players) * 100

demos_df = pd.DataFrame({"Total Player Info":[total_players],
                        "Male Players":[male],
                        "Female Players":[female],
                        "Other/Non-Disclosed":[other],
                        "Percent by Gender(Male)":[pctmale],
                        "Percent by Gender(Female)":[pctfemale],
                         "Percent by Gender(NonDisclosed)":[pctothers]})

demos_df.round(2) #this rounds all the numbers in the dataframe to 2 places past the decimal
```

```python
# Broken down by gender: Purchase Count, Avg Purchase Price, Total Purchase Value, Avg. Purchase Total Per Person by Gender
gengroup = file_one_df.groupby("Gender")

# Purchase Count
gn_count = file_one_df.groupby(["Gender"]).count()["Price"]
gn_count

# Average Purchase Value
gnaverage_price = file_one_df.groupby(["Gender"]).mean()["Price"]
gnaverage_price

# Total Purchase Value
gntotal_purch_v = file_one_df.groupby(["Gender"]).sum()["Price"]
gntotal_purch_v

  
genderspend = pd.DataFrame({"Purchase Count": gn_count,  
                      "Average Purchase Value": gnaverage_price,
                            "Total Purchase Value": gntotal_purch_v
                           })

genderspend["Average Purchase Value"] = genderspend["Average Purchase Value"].map("${:.2f}".format)
genderspend["Total Purchase Value"] = genderspend["Total Purchase Value"].map("${:.2f}".format)

genderspend
```

```python
# Broken down by 4 year age bins: Purchase Count, Avg Purchase Price, Total Purchase Value, Avg. Purchase Total Per Person by Age Group

bins = [0, 10, 15, 19, 23, 27, 31, 35, 39, 100]

# Create the names for the four bins
Age_Group_Names = ["Under 10", "10-14", "15-18", "19-22", "23-26","27-30", "31-34","35-38","39 and Up"]

#insert the data into the bins
#then group it

agedata = pd.cut(file_one_df["Age"], bins, labels = Age_Group_Names).head()

file_one_df["Age Range"] = pd.cut(file_one_df["Age"],bins,labels = Age_Group_Names)
file_one_df.head()

#Group by Age Range
group = file_one_df.groupby("Age Range")



age_ranges_df = file_one_df.groupby(["Age Range"])
age_ranges_df 


playeragecount = age_ranges_df["SN"].count() #count the screen names within the age ranges dataframe
playerpercent = (age_ranges_df["SN"].count() / total_players) * 100
playeragecount
playerpercent

agedemo_df = pd.DataFrame({"Percentage of Players": playerpercent,
                                    "Total in this Group": playeragecount
                                    })




agedemo_df.round(2)
```

```python
# Purchase Count
sn_count = file_one_df.groupby(["Age Range"]).count()["Age"]
sn_count

# Average Purchase Value
average_price = file_one_df.groupby(["Age Range"]).mean()["Price"]
average_price

# Total Purchase Value
total_purch_v = file_one_df.groupby(["Age Range"]).sum()["Price"]
total_purch_v

  
agemetrics = pd.DataFrame({"Purchase Count": sn_count, 
                      "Total Purchase Value": total_purch_v, 
                      "Average Purchase Value": average_price
                     })

agemetrics["Average Purchase Value"] = agemetrics["Average Purchase Value"].map("${:.2f}".format)
agemetrics["Total Purchase Value"] = agemetrics["Total Purchase Value"].map("${:.2f}".format)


agemetrics
```

```python
# Top 5 spenders by total purchase value: SN, Purchase Count, Avg. Purchase Price, Total Purchase Value
#Group by Screen Name
sngroup = file_one_df.groupby("SN")

# Purchase Count
sn_count = file_one_df.groupby(["SN"]).count()["Price"]
sn_count

# Average Purchase Value
average_price = file_one_df.groupby(["SN"]).mean()["Price"]
average_price

# Total Purchase Value
total_purch_v = file_one_df.groupby(["SN"]).sum()["Price"]
total_purch_v

  
spenders = pd.DataFrame({"Purchase Count": sn_count,  
                      "Average Purchase Value": average_price,
                        "Total Purchase Value": total_purch_v
                     })

spenders["Average Purchase Value"] = spenders["Average Purchase Value"].map("${:.2f}".format)
spenders["Total Purchase Value"] = spenders["Total Purchase Value"].map("${:.2f}".format)
spenders = spenders.sort_values(by="Total Purchase Value", ascending =False)

spenders
spenders.head(5)
```

```python
# 5 most profitable items by total purchase price: Item ID, Item Name, Purchase Count, Item Price, Total Purchase Value
itemcount = file_one_df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Transactions")
avgprice =file_one_df.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Avg Price")
totalprice = file_one_df.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Price")

# Convert to DataFrame

sellers_df = pd.DataFrame({"Transactions":itemcount,
                                   "Average Price":avgprice,
                                   "Total Purchase Price":totalprice,})
sellers_df = sellers_df.sort_values(by='Total Purchase Price',ascending=False)

sellers_df.head(5)
```
